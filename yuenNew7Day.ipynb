```python
from datetime import date
from datetime import datetime
from datetime import datetime, timedelta


today = date.today()
midnight = datetime.combine(today, datetime.min.time())
 
print('Midnight: %s ' % (midnight) )
```

```python

def getSaturday():
    today = midnight
    date = today.weekday()
    if date == 0:
        d = today - timedelta(days=2)
        return d
    elif date == 1:
        d = today - timedelta(days=3)
        #print(type(d))
        return d
    elif date == 2:
        d = today - timedelta(days=4)
        return d
    elif date == 3:
        d = today - timedelta(days=5)
        return d
    elif date == 4:
        d = today - timedelta(days=6)
        return d
    elif date == 5:
        d = today - timedelta(days=7)
        return d
    else:
        d = today - timedelta(days=1)
        return d
        

lastSaturday = (getSaturday())
lastSunday = getSaturday() - timedelta(days=6)

strLastSaturday = lastSaturday.strftime('%Y-%m-%d')


strLastSunday = lastSunday.strftime('%Y-%m-%d')

###########################Passing Dates for parsing into URL ###########################
```

```python
def pickCalc(x):
    if x == 'PickingPickedInTransit' or x == 'PickingPickedAtDestination' or x == "PickingPickedInProgress":
        return 'PickingPicked Total'
    elif x == 'PickingNotYetPickedPrioritized':
        return 'PickingNotYetPicked Total'
    elif x == 'PickingNotYetPicked':
        return 'PickingNotYetPicked Total'
    else: 
        return x
df_PPR['Work Pool'] = df_PPR['Work Pool'].apply(pickCalc)
df_VNA['Work Pool'] = df_VNA['Work Pool'].apply(pickCalc)
df_PPSplit['Work Pool'] = df_PPSplit['Work Pool'].apply(pickCalc)

df_PPR
```

```python
def df_manip(df, csvfilename, workpoolname):
    df['Expected Ship Datetime'] =pd.to_datetime(df['Expected Ship Datetime'])
    df['Day'] = df['Expected Ship Datetime'].dt.day_name()
    df['Day'].value_counts()
    temp = pd.DatetimeIndex(df['Expected Ship Datetime'])
    df['date'] = df['Expected Ship Datetime']
    df['date'] = temp.date
    df['time'] = temp.time
    df['hr']= temp.hour
    df['min']=temp.minute
    df['shiftHr'] = df['Day'].astype(str)+ " " + df['hr'].astype(str) + ":" + df['min'].astype(str)
    df['shiftHr'].unique()
    df=df.drop(columns=['Day', 'date', 'time'])# 'hr', 'min'])
    
    def shiftCalc(x):
        if x == 'Sunday 11:20' or x == 'Sunday 14:20' or x == 'Sunday 18:20':
            return 'FHD'
        elif x == 'Monday 11:20' or x == 'Monday 14:20' or x == 'Monday 18:20':
            return 'FHD'
        elif x == 'Tuesday 11:20' or x == 'Tuesday 14:20' or x == 'Tuesday 18:20':
            return 'FHD'
        ################ Front Half Nights
        elif x == 'Sunday 22:50' or x == 'Monday 6:20':
            return 'FHN'
        elif x == 'Monday 22:50' or x == 'Tuesday 6:20':
            return 'FHN'
        elif x == 'Tuesday 22:50' or x == 'Wednesday 6:20':
            return 'FHN'
        #######################Wednesday
        elif x == 'Wednesday 11:20' or x == 'Wednesday 14:20' or x == 'Wednesday 18:20':
            return 'WDD'
        elif x == 'Wednesday 22:50' or x == 'Thursday 6:20':
            return 'WDN'    
        ###################### BHD/BHN
        elif x == 'Thursday 11:20' or x == 'Thursday 14:20' or x == 'Thursday 18:20':
            return 'BHD'
        elif x == 'Friday 11:20' or x == 'Friday 14:20' or x == 'Friday 18:20':
            return 'BHD'
        elif x == 'Saturday 11:20' or x == 'Saturday 14:20' or x == 'Saturday 18:20':
            return 'BHD'
        #####################################
        elif x == 'Thursday 22:50' or x == 'Friday 6:20':
            return 'BHN'
        elif x == 'Friday 22:50' or x == 'Saturday 6:20':
            return 'BHN'
        elif x == 'Saturday 22:50' or x == 'Sunday 6:20':
            return 'BHN'

    df['shift'] = df['shiftHr'].apply(shiftCalc)


    
    
    df['shiftHr2'] = df['shift'].astype(str)+ " " + df['hr'].astype(str) + ":" + df['min'].astype(str)
    df = df.drop(columns = ['hr', 'min'])
    df = df.groupby(['shiftHr2', 'Work Pool'])['Quantity'].sum()
    
    df.to_csv(csvfilename)
    df2 = pd.read_csv(csvfilename)

    df3 = df2.pivot(index='Work Pool', columns='shiftHr2', values='Quantity')

    df3 = df3.fillna(0)

    df3 = df3.astype(int)
    df3.to_csv(csvfilename)
    df4 = pd.read_csv(csvfilename)#, index = 'Work Pool')
    df4 = df4.set_index('Work Pool')
    idx = pd.Index(['Packing', 'PickingNotYetPicked Total', 'PickingPicked Total', 'ProblemSolving', 'Scanned', 'Sorted'], name = "Work Pool")
    blank_df = pd.DataFrame(
        {
            'FHD 11:20':0,
            'FHD 14:20':0,
            'FHD 18:20':0, 
            'BHD 11:20':0, 
            'BHD 14:20':0, 
            'BHD 18:20':0, 
            'FHN 22:50':0, 
            'FHN 6:20':0, 
            'BHN 22:50':0, 
            'BHN 6:20':0, 
            'WDD 11:20':0, 
            'WDD 14:20':0, 
            'WDD 18:20':0, 
            'WDN 22:50':0, 
            'WDN 6:20':0
        }, index = idx)#, index = ['Packing', 'PickingNotYetPicked Total', 'PickingPicked Total', 'ProblemSolving', 'Scanned', 'Sorted'])
    #blank_df.Index.rename(name = "Work Pool")
    
    for i in df4.columns:
        blank_df = blank_df.drop(columns=[i])
    df4a = pd.concat([blank_df, df4], axis=1)
    df4a = df4a.fillna(0)
    df4a = df4a.astype(int)
    #df4a = df4a.drop(index=['ProcessPartial'])
    df4a['FHD Total'] = df4a['FHD 11:20'] + df4a['FHD 14:20'] + df4a['FHD 18:20']
    df4a['BHD Total'] = df4a['BHD 11:20'] + df4a['BHD 14:20'] + df4a['BHD 18:20']
    df4a['FHN Total'] = df4a['FHN 22:50'] + df4a['FHN 6:20']
    df4a['BHN Total'] = df4a['BHN 22:50'] + df4a['BHN 6:20']
    df4a['WDD Total'] = df4a['WDD 11:20'] + df4a['WDD 14:20'] + df4a['WDD 18:20']
    df4a['WDN Total'] = df4a['WDN 22:50'] + df4a['WDN 6:20']

    
    df4a = df4a[['FHD 11:20', 'FHD 14:20', 'FHD 18:20', 'FHD Total', 'BHD 11:20', 'BHD 14:20', 'BHD 18:20', 'BHD Total', 'FHN 22:50', 'FHN 6:20', 'FHN Total', 'BHN 22:50', 'BHN 6:20', 'BHN Total', 'WDD 11:20', 'WDD 14:20', 'WDD 18:20', 'WDD Total', 'WDN 22:50', 'WDN 6:20', 'WDN Total']]
       
    df4a.loc['Total'] = df4a.sum(numeric_only=True, axis=0)
    df4a.to_csv(csvfilename)

    df5 = pd.read_csv(csvfilename, header = 0, index_col=False)
    df5 = df5.rename(columns={"Work Pool": workpoolname})
    return df5

"""
df_PP2R = pd.DataFrame(PP2R)
df_VNA = pd.DataFrame(VNA)
df_PPSplit = pd.DataFrame(df_PPSplitDemand1)
"""

df_PP2R2 = df_manip(df_PPR, "Pick2Rebin.csv", "Pick to Rebin")
df_VNA2 = df_manip(df_VNA, "VNA.csv", "VNA")
df_PPSplit2 = df_manip(df_PPSplit, "PPSplit.csv", "PPSplitDemand")

#df_PP2R2
#df_VNA2
#df_PPSplit2

    
```

```python
import openpyxl
from openpyxl.styles import PatternFill
from openpyxl.styles import Alignment
from openpyxl.styles import Border, Side

#dir(openpyxl.load_workbook)#.workbook.workbook)9

wb = openpyxl.load_workbook("willyuenTotals1.xlsx")
ws_pprb = wb.worksheets[0]
ws_vna = wb.worksheets[1]
ws_PPSplit = wb.worksheets[2]

def excel_df(ws, processPathStr):
    
    ws.column_dimensions['A'].width = 25

    ws['A1']=processPathStr
    ws['A2']='Work Pool'
    ws['A1'].fill = PatternFill(fgColor="FFFF00", fill_type = "solid")
    ws['A1'].alignment = Alignment(horizontal='center')


    index_fill = PatternFill(fgColor='BDD7EE', fill_type = 'solid')

    ws['A2'].fill = index_fill
    ws['A3'].fill = index_fill
    ws['A4'].fill = index_fill
    ws['A5'].fill = index_fill
    ws['A6'].fill = index_fill
    ws['A7'].fill = index_fill
    ws['A8'].fill = index_fill
    ws['A9'].fill = index_fill

    ws['B2']='11:20'
    ws['C2']='14:20'
    ws['D2']='18:20'
    ws['E2']='Total'

    ws['F2']='11:20'
    ws['G2']='14:20'
    ws['H2']='18:20'
    ws['I2']='Total'

    ws['J2']='22:50'
    ws['K2']='06:20'
    ws['L2']='Total'

    ws['M2']='22:50'
    ws['N2']='06:20'
    ws['O2']='Total'

    ws['P2']='11:20'
    ws['Q2']='14:20'
    ws['R2']='18:20'
    ws['S2']='Total'

    ws['T2']='22:50'
    ws['U2']='06:20'
    ws['V2']='Total'

    ws.merge_cells(range_string="B1:E1")
    ws['B1']='FHDs'
    ws['B1'].alignment = Alignment(horizontal='center')
    FHDs_fill = PatternFill(fgColor='AFDDD5', fill_type = 'solid')

    ws['B1'].fill = FHDs_fill
    ws['B2'].fill = FHDs_fill
    ws['C2'].fill = FHDs_fill
    ws['D2'].fill = FHDs_fill
    ws['E2'].fill = FHDs_fill

    ws.merge_cells(range_string="F1:I1")
    ws['F1']='BHDs'
    ws['F1'].alignment = Alignment(horizontal='center')
    BHDs_fill = PatternFill(fgColor='C6E0B4', fill_type = 'solid')

    ws['F1'].fill = BHDs_fill
    ws['F2'].fill = BHDs_fill
    ws['G2'].fill = BHDs_fill
    ws['H2'].fill = BHDs_fill
    ws['I2'].fill = BHDs_fill

    ws.merge_cells(range_string="J1:L1")
    ws['J1']='FHNs'
    ws['J1'].alignment = Alignment(horizontal='center')
    FHNs_fill = PatternFill(fgColor='F8CBAD', fill_type = 'solid')

    ws['J1'].fill = FHNs_fill
    ws['J2'].fill = FHNs_fill
    ws['K2'].fill = FHNs_fill
    ws['L2'].fill = FHNs_fill


    ws.merge_cells(range_string="M1:O1")
    ws['M1']='BHNs'
    ws['M1'].alignment = Alignment(horizontal='center')
    BHNs_fill = PatternFill(fgColor='FFE699', fill_type = 'solid')

    ws['M1'].fill = BHNs_fill
    ws['M2'].fill = BHNs_fill
    ws['N2'].fill = BHNs_fill
    ws['O2'].fill = BHNs_fill

    ws.merge_cells(range_string="P1:S1")
    ws['P1']='Wednesday Day'
    WDD_fill = PatternFill(fgColor='85E3FF', fill_type = 'solid')

    ws['P1'].fill = WDD_fill
    ws['P2'].fill = WDD_fill
    ws['Q2'].fill = WDD_fill
    ws['R2'].fill = WDD_fill
    ws['S2'].fill = WDD_fill

    ws['P1'].alignment = Alignment(horizontal='center')

    ws.merge_cells(range_string="T1:V1")
    ws['T1']='Wednesday Night'
    ws['T1'].alignment = Alignment(horizontal='center')
    WDN_fill = PatternFill(fgColor='DCD3FF', fill_type = 'solid')

    ws['T1'].fill = WDN_fill
    ws['T2'].fill = WDN_fill
    ws['U2'].fill = WDN_fill
    ws['V2'].fill = WDN_fill
    
    
    borderStyle = Border(top = Side(border_style='thin', color='FF000000'),    
                              right = Side(border_style='thin', color='FF000000'), 
                              bottom = Side(border_style='thin', color='FF000000'),
                              left = Side(border_style='thin', color='FF000000'))
    range = 'A1:V9'
    def set_border(ws, cell_range):
        rows = ws[cell_range]
        side = Side(border_style='thin', color="FF000000")

        rows = list(rows)  # we convert iterator to list for simplicity, but it's not memory efficient solution
        max_y = len(rows) - 1  # index of the last row
        for pos_y, cells in enumerate(rows):
            max_x = len(cells) - 1  # index of the last cell
            for pos_x, cell in enumerate(cells):
                border = Border(
                    left=cell.border.left,
                    right=cell.border.right,
                    top=cell.border.top,
                    bottom=cell.border.bottom
                )
                if pos_x == 0:
                    border.left = side
                if pos_x == max_x:
                    border.right = side
                if pos_y == 0:
                    border.top = side
                if pos_y == max_y:
                    border.bottom = side

                # set new border only if it's one of the edge cells
                if pos_x == 0 or pos_x == max_x or pos_y == 0 or pos_y == max_y:
                    cell.border = border
    set_border(ws, "A1:V9")
    set_border(ws, "B3:B9")
    set_border(ws, "C3:C9")
    set_border(ws, "D3:D9")
    set_border(ws, "E3:E9")
    set_border(ws, "F3:F9")
    set_border(ws, "G3:G9")
    set_border(ws, "H3:H9")
    set_border(ws, "I3:I9")
    set_border(ws, "J3:J9")
    set_border(ws, "K3:K9")
    set_border(ws, "L3:L9")
    set_border(ws, "M3:M9")
    set_border(ws, "N3:N9")
    set_border(ws, "O3:O9")
    set_border(ws, "P3:P9")
    set_border(ws, "Q3:Q9")
    set_border(ws, "R3:R9")
    set_border(ws, "S3:S9")
    set_border(ws, "T3:T9")
    set_border(ws, "U3:U9")
    set_border(ws, "V3:V9")
    
    set_border(ws, "B1:E1")
    set_border(ws, "F3:I1")
    set_border(ws, "J1:L1")
    set_border(ws, "M1:O1")
    set_border(ws, "P1:S1")
    set_border(ws, "T1:V1")
    
    set_border(ws, "B3:V3")
    set_border(ws, "B4:V4")
    set_border(ws, "B5:V5")
    set_border(ws, "B6:V6")
    set_border(ws, "B7:V7")
    set_border(ws, "B8:V8")
    set_border(ws, "B9:V9")
    
    

    
    ws['B1'].border = borderStyle

    ws['F1'].border = borderStyle
    ws['J1'].border = borderStyle
    ws['M1'].border = borderStyle
    ws['P1'].border = borderStyle
    ws['T1'].border = borderStyle
    
    ws['A3'].border = borderStyle
    ws['A4'].border = borderStyle
    ws['A5'].border = borderStyle
    ws['A6'].border = borderStyle
    ws['A7'].border = borderStyle
    ws['A8'].border = borderStyle
    ws['A9'].border = borderStyle
    
    
    


    wb.save("willyuenTotals1.xlsx")

excel_df(ws_pprb, "Pick To Rebin")
excel_df(ws_vna, "VNA")
excel_df(ws_PPSplit, "PPSplitDemand")
```
